```python
# Cargar a la memoria de trabajo las bibliotecas: Pandas, 
# Numpy y la biblioteca de gr√°ficas Matplotlib
# Se asignan nombres cortos.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfDec = pd.read_csv('ValentiaDec.txt', skiprows=3, sep='\s+')
dfJun = pd.read_csv('ValentiaJun.txt', skiprows=3, sep='\s+')
dfDec.columns = ['PRES', 'HGHT', 'TEMP', 'DWPT', 'RELH', 'MIXR', 'DRCT', 'SKNT', 'THTA', 'THTE', 'THTV']
dfJun.columns = ['PRES', 'HGHT', 'TEMP', 'DWPT', 'RELH', 'MIXR', 'DRCT', 'SKNT', 'THTA', 'THTE', 'THTV']
```

```python
dfDec = dfDec.drop([0])
dfJun = dfJun.drop([0])
dfDec = dfDec[:-26]
dfJun = dfJun[:-26]
```

```python
col = ['PRES', 'HGHT', 'TEMP', 'DWPT', 'RELH', 'MIXR', 'DRCT', 'SKNT', 'THTA', 'THTE', 'THTV']
dfDec[col] = dfDec[col].apply(pd.to_numeric, errors='coerce', axis=1)
dfJun[col] = dfJun[col].apply(pd.to_numeric, errors='coerce', axis=1)
```

```python
#Gr√°fica de la variaci√≥n de la presi√≥n con respecto a la altura.

#Datos de 22 de Diciembre 2017, Valentia Observatory.
presD = dfDec['PRES']
altD = dfDec['HGHT']
plt.plot(presD, altD, "blue", label='PRES'); plt.legend(loc='best')
plt.title('Presi√≥n con respecto a la Altura - Diciembre')
plt.ylabel('Altura (m)')
plt.xlabel('Presi√≥n (hPa)')
plt.grid(True)
plt.show()

#Datos de 22 de Junio 2017, Valentia Observatory.
presJ = dfJun['PRES']
altJ = dfJun['HGHT']
plt.plot(presJ, altJ, "red", label='PRES'); plt.legend(loc='best')
plt.title('Presi√≥n con respecto a la Altura - Junio')
plt.ylabel('Altura (m)')
plt.xlabel('Presi√≥n (hPa)')
plt.grid(True)
plt.show()
```

```python
#Gr√°fica de la varciaci√≥n de la temperatura con respecto a la altura.

#Datos de 22 de Diciembre 2917, Valentia Observatory.
tempD = dfDec['TEMP']
altD = dfDec['HGHT']
plt.plot(tempD, altD, "blue", label='TEMP'); plt.legend(loc='best')
plt.title('Temperatura con respecto a la Altura - Diciembre')
plt.ylabel('Altura (m)')
plt.xlabel('Temperatura (C)')
plt.grid(True)
plt.show()

#Datos de 22 de Junio 2017, Valentia Observatory.
tempJ = dfJun['TEMP']
altJ = dfJun['HGHT']
plt.plot(tempJ, altJ, "red", label='TEMP'); plt.legend(loc='best')
plt.title('Temperatura con respecto a la Alutra - Junio')
plt.ylabel('Altura (m)')
plt.xlabel('Temperatura (C)')
plt.grid(True)
plt.show()
```

```python
#Gr√°fica de la temperatura (TEMP) y temperatura de roc√≠o (DEWPT), como funci√≥n de la altura.

#Datos de 22 de Diciembre 2917, Valentia Observatory
tempD = dfDec['TEMP']
dwptD = dfDec['DWPT']
altD = dfDec['HGHT']
plt.plot(tempD, altD, "blue", label='TEMP')
plt.plot(dwptD, altD, "cyan", label='DWPT'); plt.legend(loc='best')
plt.title('Temperatura y Temperatura de Roc√≠o con respecto a la Altura - Diciembre')
plt.ylabel('Altura (m)')
plt.xlabel('Temperatura (C)')
plt.grid(True)
plt.show()

#Datos de 22 de Diciembre 2917, Valentia Observatory
tempJ = dfJun['TEMP']
dwptJ = dfJun['DWPT']
altJ = dfJun['HGHT']
plt.plot(tempJ, altJ, "red", label='TEMP')
plt.plot(dwptJ, altJ, "orange", label='DWPT'); plt.legend(loc='best')
plt.title('Temperatura y Temperatura de Roc√≠o con respecto a la Altura - Junio')
plt.ylabel('Altura (m)')
plt.xlabel('Temperatura (C)')
plt.grid(True)
plt.show()
```

```python
#Gr√°fica de la rapidez de los vientos en nudos.

#Datos de 22 de Diciembre 2017, Valentia Observatory.
rapD = dfDec['SKNT']
altD = dfDec['HGHT']
plt.plot(rapD, altD, "blue", label='SKNT'); plt.legend(loc='best')
plt.title("Rapidez de los Vientos - Diciembre")
plt.ylabel("Altura (m)")
plt.xlabel("Rapidez (nudos)")
plt.xlim([0,80])
plt.ylim([0,30000])
plt.grid(True)
plt.show()

#Datos de 22 de Junio 2017, Valentia Observatory.
rapJ = dfJun['SKNT']
altJ = dfJun['HGHT']
plt.title("Rapidez de los Vientos - Junio")
plt.plot(rapJ, altJ, "red", label='SKNT'); plt.legend(loc='best')
plt.ylabel("Altura (m)")
plt.xlabel("Rapidez (nudos)")
plt.xlim([0,75])
plt.ylim([0,30000])
plt.grid(True)
plt.show()
```

```python
#Gr√°fica de la humedad relativa como funci√≥n de la altura.  

#Datos de 22 de Diciembre 2017, Valentia Observatory.
humD = dfDec['RELH']
altD = dfDec['HGHT']
plt.plot(humD, altD, "blue", label='RELH'); plt.legend(loc='best')
plt.title('Humedad Relativa con respecto a la Altura - Diciembre')
plt.xlabel('Humedad Relativa (%)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()

#Datos de 22 de Junio 2017, Valentia Observatory.
humJ = dfJun['RELH']
altJ = dfJun['HGHT']
plt.plot(humJ, altJ, "red", label='RELH'); plt.legend(loc='best')
plt.title('Humedad Relativa con respecto a la Altura - Junio')
plt.xlabel('Humedad Relativa (%)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()
```
